# DEPRECATED PART OF EDGAR

```python
#THIS API ENDPOINT DOESNT APPEAR TO HAVE MUCH FRESH DATA!

yr = range(2019,2023)
qtr = range(1,4)

d = {}
ftbl = pd.DataFrame()
for y in yr:
    for q in qtr:
        f = edgar.get_frames(taxonomy="us-gaap", tag="NetIncomeLoss", unit="USD", year=str(y), quarter=str(q))
        for ent in f['data']:
            
            if not ent['cik'] in d:
                d[ent['cik']] = []
            
            d[ent['cik']].append([str(y)+'Q'+str(q),ent['val']])

for k in d.keys():
    df = pd.DataFrame()
    df['cik'] = ''
    df.at[0,'cik'] =k
    r = d[k]
    for qt in r:
        df[qt[0]] = 0
        df.at[0, qt[0]] = int(qt[1])

    ftbl = pd.concat([ftbl,df],axis=0) 

#edgar.get_frames(taxonomy="us-gaap", tag="EarningsPerShareDilute", unit="USD", year="2019", quarter=1)
```
